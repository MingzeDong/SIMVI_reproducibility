```python
import scanpy as sc
import squidpy as sq

import numpy as np
import pandas as pd
```

```python
adata_mtg = sc.read_h5ad('mtg_baselines2.h5ad')
sc.pp.normalize_total(adata_mtg)
sc.pp.log1p(adata_mtg)
sc.pp.highly_variable_genes(adata_mtg,n_top_genes=1000)

adata_mtg.obsm['spatial'][adata_mtg.obs['batch']==adata_mtg.obs['batch'].unique()[0]] -= 5000
adata_mtg.obsm['spatial'][adata_mtg.obs['batch']==adata_mtg.obs['batch'].unique()[2]] += 5000

sc.pl.spatial(adata_mtg,color=['cluster_L1'],spot_size = 30)
```

```python
from GraphST.GraphST import GraphST

for i in range(10):
    model = GraphST(adata_mtg, device='cuda:1',random_seed=i,datatype='Slide')
    adata_mtg.obsm['GraphST_'+str(i)] = model.train()
```
